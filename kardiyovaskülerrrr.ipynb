```python
import pandas as pd

# Dosya yolunu kendi dosyana göre güncelle
file_path = '/content/drive/MyDrive/cardio.csv'
df = pd.read_csv(file_path, sep=';')

# İlk 5 satırı görelim
df.head()
```

```python
#VERİ YAPISINI TANIMA ÖZETLEME, EKSİK DEĞER KONTROLÜ


# Genel bilgi
df.info()

# İstatistiksel özet
df.describe()

# Eksik değer kontrolü
df.isnull().sum()





#READ ME
"""
  70.000 birey/hasta verisi mevcut.

Toplam 13 özellik (kolon) var: id, age, gender, height, weight, vs.
Türleri çoğunlukla int64 → tam sayı.
weight sütunu float64 → ondalıklı değer içeriyor.
Tüm sütunlar eksiksiz (null yok) → Temiz veri seti!
Hafif bir veri seti, RAM dostu. Yani makineyi zorlamaz.
Burada her sütun için eksik değer sayısı = 0 olarak gösterildi.



"""
```

```python
#YAŞ DEĞERİNİ GÜN CİNSİNDEN YAZDIRIYOR

df['age'] = (df['age'] / 365).astype(int)
```

```python
"""
sadece şu 4 sayısal sütun için temel istatistikleri çıkarır:

ap_hi → Sistolik tansiyon (büyük tansiyon)

ap_lo → Diastolik tansiyon (küçük tansiyon)

height → Boy (cm)

weight → Kilo (kg)

"""

df[['ap_hi', 'ap_lo', 'height', 'weight']].describe()






#read me
"""
Bu satırda ap_hi, ap_lo, height, weight değişkenleri için istatistiksel özet çıkarılıyor.

Kısaca sütunlar:
ap_hi: Sistolik tansiyon (büyük tansiyon)

ap_lo: Diastolik tansiyon (küçük tansiyon)

height: Boy (cm)

weight: Kilo (kg




count	Gözlem sayısı → 70.000
mean	Ortalama
std	Standart sapma (dağılım genişliği)
min	En küçük değer
25%, 50%, 75%	Çeyrekler (Q1, Medyan, Q3)
max	En büyük değer

"""
```

```python
#Her satır, ilgili sütundaki mantıksız veya aşırı uç değerleri filtreleyerek veri setini temizliyor.





df = df[(df['ap_hi'] > 80) & (df['ap_hi'] < 200)]
df = df[(df['ap_lo'] > 40) & (df['ap_lo'] < 150)]
df = df[(df['height'] > 100) & (df['height'] < 250)]
df = df[(df['weight'] > 30) & (df['weight'] < 200)]
```

```python
#Verileri modellemeye hazırlarken yapılan ölçekleme (standardizasyon) işlemini içeriyor



from sklearn.preprocessing import StandardScaler

features = ['age', 'gender', 'height', 'weight', 'ap_hi', 'ap_lo',
            'cholesterol', 'gluc', 'smoke', 'alco', 'active']

X = df[features]

# Ölçekleme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#K-Means kümeleme algoritmasında en önemli kararlardan biri olan optimal küme sayısını (K) belirlemek için kullanılan Elbow (Dirsek) Yöntemini



from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,4))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Küme Sayısı (k)')
plt.ylabel('İnertia')
plt.title('Elbow Yöntemi ile Optimal K Belirleme')
plt.grid()
plt.show()





"""



Grafiğe göre:

K = 3 ile K = 4 arasında “dirsek” oluşmuş.

Bu noktadan sonra azalma daha yavaş, yani eklenen küme eskisi kadar iyileştirme sağlamıyor.


İnertia'da büyük bir düşüş sağlanmış,

Ama K > 3’ten sonra maliyet artıyor, fayda azalıyor.



"""
```

```python
#K-Means kümeleme modelinin eğitimini ve çıktılarının veri setine eklenmesini sağlıyor



# K-Means modelini oluştur ve eğit
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Sonuçları veri çerçevemize ekleyelim
df['cluster'] = clusters

```

```python

#PCA + K-Means görselleştirme kodu, karmaşık (çok boyutlu) veriyi sadeleştirip, kümeleri insan gözüyle anlaşılır hale getirmek için kullanılır.





from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', alpha=0.6)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('PCA ile Kümeleme Görselleştirme')
plt.colorbar(label='Küme Numarası')
plt.grid(True)
plt.show()






"""




 PCA (Principal Component Analysis)
Yüksek boyutlu veriyi (senin verin 11 özellikliydi) 2 boyuta indiriyor.

Böylece kümeler görsel olarak daha kolay anlaşılır hale geliyor.

PCA 1 ve PCA 2, yeni oluşturulan eksenlerdir (verinin en çok bilgi içeren yönleri).




Her bir nokta bir bireyi (hastayı) temsil ediyor.

Renkler → K-Means ile belirlenen küme numaraları (0, 1, 2).

Benzer renkler gruplanmışsa → algoritma benzer kişileri doğru gruplayabilmiş.

Küme ayrımı net görünüyorsa → K-Means iyi çalışmış demektir.



"""
```

```python
#K-Means kümeleme sonrası oluşan her kümenin istatistiksel özetini çıkarır.


cluster_summary = df.groupby('cluster')[features].mean().round(2)
cluster_summary
```

```python
#DBSCAN algoritması ile yoğunluk tabanlı kümeleme yapar ve aynı zamanda aykırı (uç) değerleri tespit eder.




from sklearn.cluster import DBSCAN

# DBSCAN modelini oluştur
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_clusters = dbscan.fit_predict(X_scaled)

# Aykırı değerleri (outlier) tespit et
df['dbscan_cluster'] = dbscan_clusters

# Aykırı veriler, -1 olarak işaretlenir
outliers = df[df['dbscan_cluster'] == -1]

# Aykırı verileri inceleyelim
outliers.head()


#-1 OLMASI AYKIRI DEĞERİ GÖSTERİR ÖRN: BOY
```
